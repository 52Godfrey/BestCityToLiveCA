```python
#Dependencies:
    
import pandas as pd
import os
```

```python
# Path to file folder
data_folder_own = 'Resources/RentData/Owning Only/'
data_folder_rent = 'Resources/RentData/Rent only/'
```

```python
# Get list of files from given directory
files_own = os.listdir(data_folder_own)
files_rent = os.listdir(data_folder_rent)
```

### Data Preparation

```python
# DataFrame to store data about all homes type
homes_df = pd.DataFrame()

# Read each file to a temp DataFrame, drop excess columns, and concatenates temp DataFrame with homes_df
# Also we add column Home Type to identify home type in homes_df
for ind in range(len(files_own) - 1):
    path_to_file = data_folder_own + files_own[ind]
    temp_df = pd.read_csv(path_to_file)
    temp_df = temp_df.filter(regex=r'RegionID|SizeRank|RegionName|RegionType|State|StateCodeFIPS|MunicipalCodeFIPS|2019')
    temp_df['Home Type'] = files_own[ind]
    homes_df = pd.concat([homes_df, temp_df], ignore_index=True)

homes_df
```

```python
# Change value from file names to home type

homes_df.loc[homes_df['Home Type'] == '4_bedroom.csv', 'Home Type'] = '4 bedroom'
homes_df.loc[homes_df['Home Type'] == 'Single_Family_homes.csv', 'Home Type'] = 'single family'
homes_df.loc[homes_df['Home Type'] == '1_bedroom.csv', 'Home Type'] = '1 bedroom'
homes_df.loc[homes_df['Home Type'] == '2_bedroom.csv', 'Home Type'] = '2 bedroom'
homes_df.loc[homes_df['Home Type'] == 'Condo_Co-op.csv', 'Home Type'] = 'condo'
homes_df.loc[homes_df['Home Type'] == '5_bedroom.csv', 'Home Type'] = '5 bedroom'
```

```python
# final data for homes 
homes_ca_df = homes_df[homes_df['StateName'] == 'CA']
homes_ca_df['Home Type'].unique()
```

```python
# add new column "Avg house price"

homes_ca_df['Avg house price'] = round((homes_ca_df['2019-01-31'] + homes_ca_df['2019-02-28']
                                        + homes_ca_df['2019-03-31'] + homes_ca_df['2019-04-30'] 
                                        + homes_ca_df['2019-05-31'] + homes_ca_df['2019-06-30'] 
                                        + homes_ca_df['2019-07-31'] + homes_ca_df['2019-08-31'] 
                                        + homes_ca_df['2019-09-30'] + homes_ca_df['2019-10-31'] 
                                        + homes_ca_df['2019-11-30'] + homes_ca_df['2019-12-31']) / 12, 2)
```

```python
# choose only column that we need for analisys
home_ca_value_df = homes_ca_df[['RegionName', 'State', 'Avg house price', 'Home Type']].copy()
```

```python
# get population data
path_to_file = 'Resources/CrimeData/AllCountyPopulation.csv'
population_df = pd.read_csv(path_to_file)
```

```python
#Extract county name
county = []
county_names = []

for index, row in home_ca_value_df.iterrows():
    reg = row['RegionName'].split(' ')
    if len(reg) > 1:
        county.append(reg[: len(reg)-1])
    elif len(reg) == 1:
        county.append(reg)

for ind in range(len(county)):
    strg = ' '.join(county[ind])
    county_names.append(strg)
```

```python
home_ca_value_df['County'] = county_names
```

```python
home_ca_value_df = home_ca_value_df.drop(columns=['RegionName'])
clean_home_data = home_ca_value_df[['County', 'State', 'Home Type', 'Avg house price']]
clean_home_data
```

```python
# Save df to csv file
clean_home_data[['County', 'State', 'Home Type', 'Avg house price']].to_csv('Resources/HomeValue_clean.csv', index=False)
```
