Dependencies:

```python
import pandas as pd
import os
```

```python
# Path to file folder
data_folder_own = 'Resources/RentData/Owning Only/'
data_folder_rent = 'Resources/RentData/Rent only/'
```

```python
# Get list of files from given directory
files_own = os.listdir(data_folder_own)
files_rent = os.listdir(data_folder_rent)
```

### Data Preparation

```python
# DataFrame to store data about all homes type
homes_df = pd.DataFrame()

# Read each file to a temp DataFrame, drop excess columns, and concatenates temp DataFrame with homes_df
# Also we add column Home Type to identify home type in homes_df
for ind in range(len(files_own) - 1):
    path_to_file = data_folder_own + files_own[ind]
    temp_df = pd.read_csv(path_to_file)
    temp_df = temp_df.filter(regex=r'RegionID|SizeRank|RegionName|RegionType|State|StateCodeFIPS|MunicipalCodeFIPS|2019')
    temp_df['Home Type'] = files_own[ind]
    homes_df = pd.concat([homes_df, temp_df], ignore_index=True)

homes_df
```

```python
# Change value from file names to home type

homes_df.loc[homes_df['Home Type'] == '4_bedroom.csv', 'Home Type'] = '4 bedroom'
homes_df.loc[homes_df['Home Type'] == 'Single_Family_homes.csv', 'Home Type'] = 'single family'
homes_df.loc[homes_df['Home Type'] == '1_bedroom.csv', 'Home Type'] = '1_bedroom.csv'
homes_df.loc[homes_df['Home Type'] == '2_bedroom.csv', 'Home Type'] = '2_bedroom.csv'
homes_df.loc[homes_df['Home Type'] == 'Condo_Co-op.csv', 'Home Type'] = 'condo'
homes_df.loc[homes_df['Home Type'] == '5_bedroom.csv', 'Home Type'] = '5 bedroom'
```

```python
# final data for homes 
homes_df = homes_df[homes_df['StateName'] == 'CA']
homes_df
```
